# The Universal Approximation Theorem

Is the future of AI as bright as it seems right now?

Lets look at some foundational issues first.

The universal approximation theorem states that neural networks can approximate any function.

There are two important facts to be noted.

1. It is important to realize that the word used is approximate. This means that the function computed is not exact. However, this approximation can be improved as the number of computation units i.e. neurons are increased in the layer and it can be fit to the desired accuracy.

2. The function we compute this way are continuous functions. The generality does not hold for discontinuous functions. However, more often than not, continuous functions prove to be a good enough substitute for discontinuous functions, and therefore, the neural networks can work.

# Limitations of Deep Learning Models

1. **Adversarial Examples** - "If you put garbage in you will get garbage out". It is possible to do gradient ascent in input space to generate inputs that maximize the activation of some convnet filter when doing image classification.  Through gradient ascent, one can slightly modify an image in order to maximize the class prediction for a given class. By taking a picture of a panda and adding to it a "gibbon" gradient, we can get a neural network to classify this panda as a gibbon. This evidences both the brittleness of these models, and the deep difference between the input-to-output mapping that they operate and our own human perception. In short, deep learning models do not have any understanding of their input, at least not in any human sense.

2. **Local Generalization Vs. Extreme Generalization** - What deep nets do can be called "local generalization": the mapping from inputs to outputs performed by deep NNs quickly stops making sense if new inputs differ even slightly from what they saw at training time. Consider, for instance, the problem of learning the appropriate launch parameters to get a rocket to land on the moon. If you were to use a deep NN for this task, whether training using supervised learning or reinforcement learning, you would need to feed it with thousands or even millions of launch trials, i.e. you would need to expose it to a dense sampling of the input space, in order to learn a reliable mapping from input space to output space. By contrast, humans can use their power of abstraction to come up with physical models—rocket science—and derive an exact solution that will get the rocket on the moon in just one or few trials. Similarly, if you developed a deep NN controlling a human body, and wanted it to learn to safely navigate a city without getting hit by cars, the NN would have to die many thousands of times in various situations until it could infer that cars and dangerous, and develop appropriate avoidance behaviors. Dropped into a new city, the NN would have to relearn most of what it knows. On the other hand, humans are able to learn safe behaviors without having to die even once—again, thanks to their power of abstract modeling of hypothetical situations.

Empirical studies suggest that machine learning models often rely on features, such as the background, that may be spuriously correlated with the label only during training time, resulting in poor accuracy during test-time. [Link](https://arxiv.org/abs/2010.15775)

The only real success of deep learning so far has been the ability to map space X to space Y using a continuous geometric transform, given large amounts of human-annotated data. Doing this well is a game-changer for essentially every industry, but it is still a very long way from human-level AI.

# Uncertainty in deep learning models

With recent advancements in ML and AI, Deep Learning models are applied to real-world. There could be scenarios when such automated systems can be life threatening for humans.These
include automated decision making or recommendation systems in the medical domain,
autonomous control of drones and self driving cars etc.

We need tools to model confidence in AI and ML systems. For example, lets assume we have a CNN based model that classifies oranges and apples. The output layer gives a probability value where a Probability greater than 0.5 is considered to be the outcome for apple otherwise its an orange. The input can have noise (pictures of orange and apples together). In such instances of uncertainty we want our model to have some level of confidence in giving us the predictions.

If we input a picture of a watermelon (out of sample/distribution), most likely the model will not recognize it but it will give some output which will either be closer to apple or orange. This is epistemic uncertainty and points to out of distribution uncertainty.

Even though modern deep learning models used in practice do not capture model
confidence, they are closely related to a family of probabilistic models which induce
probability distributions over functions: the Gaussian process. Given a neural network,
by placing a probability distribution over each weight (a standard normal distribution for
example), a Gaussian process can be recovered in the limit of infinitely many weights. For a finite number of weights, model uncertainty can still
be obtained by placing distributions over the weights—these models are called Bayesian
neural networks.

# Types of Uncertainties


1. Epistemic: Model based uncertainty, describes confidence of the prediction and is very high when there is missing training data. It is more challenging and can be reduced to some extent by adding more data.

2. Aleotoric: Stems from data and can be considered as irreducible noise. Describes the confidence in the input data. It can be learned usinf NNs.
